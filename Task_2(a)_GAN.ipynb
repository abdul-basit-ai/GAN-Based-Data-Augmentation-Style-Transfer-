```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.utils as vutils
import matplotlib.pyplot as plt
import numpy as np
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# -------------------------
# Generator (CNN Encoder-Decoder)
# -------------------------
class Generator(nn.Module):
    def __init__(self, nz=100):
        super(Generator, self).__init__()
        self.net = nn.Sequential(
            nn.ConvTranspose2d(nz, 1024, 4, 1, 0, bias=False),     # 1x1 → 4x4
            nn.BatchNorm2d(1024),
            nn.ReLU(True),

            nn.ConvTranspose2d(1024, 512, 4, 2, 1, bias=False),    # 4x4 → 8x8
            nn.BatchNorm2d(512),
            nn.ReLU(True),

            nn.ConvTranspose2d(512, 256, 4, 2, 1, bias=False),     # 8x8 → 16x16
            nn.BatchNorm2d(256),
            nn.ReLU(True),

            nn.ConvTranspose2d(256, 128, 4, 2, 1, bias=False),     # 16x16 → 32x32
            nn.BatchNorm2d(128),
            nn.ReLU(True),

            nn.ConvTranspose2d(128, 64, 4, 2, 1, bias=False),      # 32x32 → 64x64
            nn.BatchNorm2d(64),
            nn.ReLU(True),

            nn.ConvTranspose2d(64, 32, 4, 2, 1, bias=False),       # 64x64 → 128x128
            nn.BatchNorm2d(32),
            nn.ReLU(True),

            nn.ConvTranspose2d(32, 3, 4, 2, 1, bias=False),        # 128x128 → 256x256
            nn.Sigmoid()
        )

    def forward(self, z):
        return self.net(z)

# -------------------------
# Discriminator (CNN)
# -------------------------
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.net = nn.Sequential(
            nn.Conv2d(3, 32, 4, 2, 1),        # 256 → 128
            nn.LeakyReLU(0.2, inplace=True),

            nn.Conv2d(32, 64, 4, 2, 1),       # 128 → 64
            nn.BatchNorm2d(64),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Conv2d(64, 128, 4, 2, 1),      # 64 → 32
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Conv2d(128, 256, 4, 2, 1),     # 32 → 16
            nn.BatchNorm2d(256),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Conv2d(256, 512, 4, 2, 1),     # 16 → 8
            nn.BatchNorm2d(512),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Conv2d(512, 1024, 4, 2, 1),    # 8 → 4
            nn.BatchNorm2d(1024),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Conv2d(1024, 1, 4, 1, 0),      # 4x4 → 1x1
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.net(x).view(-1)

# -------------------------
# Hyperparameters
# -------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
nz = 100
lrD = 0.0002
lrG = 0.0001
lr = 0.0002
num_epochs = 500
batch_size = 8

G = Generator(nz).to(device)
D = Discriminator().to(device)

criterion = nn.BCELoss()
# optimizerG = optim.Adam(G.parameters(), lr=lrG, betas=(0.5, 0.999))
# optimizerD = optim.Adam(D.parameters(), lr=lrD, betas=(0.5, 0.999))

optimizerG = optim.Adam(G.parameters(), lr=lr, betas=(0.5, 0.999))
optimizerD = optim.Adam(D.parameters(), lr=lr, betas=(0.5, 0.999))

# -------------------------
# Data Loading
# -------------------------
transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(256),
    transforms.ToTensor() # Normalize to [-1, 1]
])

train_dataset = datasets.ImageFolder(root="/kaggle/input/vision-project-data/Newdata/Train", transform=transform)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

# -------------------------
# Tracking
# -------------------------
G_losses = []
D_losses = []
fixed_noise = torch.randn(8, nz, 1, 1, device=device)

# -------------------------
# Training Loop
# -------------------------
for epoch in range(num_epochs):
    for i, (real_images, _) in enumerate(train_loader):
        real_images = real_images.to(device)
        batch_size = real_images.size(0)

        # Real and fake labels
        real_labels = torch.ones(batch_size, device=device)
        fake_labels = torch.zeros(batch_size, device=device)

        # ---- Train Discriminator ----
        D.zero_grad()
        output_real = D(real_images)
        loss_real = criterion(output_real, real_labels)

        noise = torch.randn(batch_size, nz, 1, 1, device=device)
        fake_images = G(noise)
        output_fake = D(fake_images.detach())
        loss_fake = criterion(output_fake, fake_labels)

        loss_D = loss_real + loss_fake
        loss_D.backward()
        optimizerD.step()

        # ---- Train Generator ----
        G.zero_grad()
        output = D(fake_images)
        loss_G = criterion(output, real_labels)
        loss_G.backward()
        optimizerG.step()

        # Store losses
        G_losses.append(loss_G.item())
        D_losses.append(loss_D.item())

        # Log
        if i % 50 == 0:
            print(f"Epoch [{epoch+1}/{num_epochs}] Batch {i}/{len(train_loader)} "
                  f"Loss D: {loss_D.item():.4f}, Loss G: {loss_G.item():.4f}")

    # ---- Visualization per epoch ----
    with torch.no_grad():
        fake = G(fixed_noise).detach().cpu()
    grid_fake = vutils.make_grid(fake, normalize=True)
    grid_real = vutils.make_grid(real_images[:8].cpu(), normalize=True)

    # plt.figure(figsize=(12, 6))
    # plt.subplot(1, 2, 1)
    # plt.title("Real Images")
    # plt.imshow(np.transpose(grid_real, (1, 2, 0)))
    # plt.axis("off")

    # plt.subplot(1, 2, 2)
    # plt.title(f"Fake Images at Epoch {epoch+1}")
    # plt.imshow(np.transpose(grid_fake, (1, 2, 0)))
    # plt.axis("off")
    # plt.show()

    # plt.plot()
    # plt.title(f"Fake Images at Epoch {epoch+1}")
    # plt.imshow(np.transpose(grid_fake, (1, 2, 0)))
    # plt.axis("off")
    # plt.show()
    # ---- Visualization per epoch ----
    if (epoch + 1) % 50 == 0:  # Display every 50 epochs
        with torch.no_grad():
            fake = G(fixed_noise).detach().cpu()
        grid_fake = vutils.make_grid(fake, normalize=True)
    
        plt.figure(figsize=(6, 6))
        plt.title(f"Fake Images at Epoch {epoch+1}")
        plt.imshow(np.transpose(grid_fake, (1, 2, 0)))
        plt.axis("off")
        plt.show()

# -------------------------
# Final Loss Plot
# -------------------------
plt.figure(figsize=(10, 10))
plt.title("Generator and Discriminator Loss During Training")
plt.plot(G_losses, label="G")
plt.plot(D_losses, label="D")
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.legend()
plt.show()


# -------------------------
# Generate New Fake Images After Training (no normalization needed)
# -------------------------
import os
from PIL import Image

output_dir = "./generated_images"
os.makedirs(output_dir, exist_ok=True)

num_images = 10
G.eval()
with torch.no_grad():
    for idx in range(num_images):
        noise = torch.randn(1, nz, 1, 1, device=device)
        fake_image = G(noise).detach().cpu()
        fake_image = fake_image.squeeze(0)  # remove batch dimension
        fake_image = transforms.ToPILImage()(fake_image)  # no denormalization required
        
        image_path = os.path.join(output_dir, f"fake_{idx+1}.png")
        fake_image.save(image_path)
        display(fake_image)
```
