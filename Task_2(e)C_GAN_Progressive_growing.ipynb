```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
from torchvision import datasets
import matplotlib.pyplot as plt
import torchvision.utils as vutils

# === Dataset Loading ===
transform = transforms.Compose([transforms.ToTensor()])
train_dataset = datasets.ImageFolder(root="/kaggle/input/vision-project-data/Newdata/Train", transform=transform)
test_dataset = datasets.ImageFolder(root="/kaggle/input/vision-project-data/Newdata/Test", transform=transform)
train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=8, shuffle=False)

# === Hyperparameters ===
latent_dim = 10
image_size = 256
batch_size = 32
epochs = 300
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# === Track Losses ===
d_losses = []
g_losses = []

# === Helper Functions ===
def conv_block(in_channels, out_channels, kernel_size=3, stride=2, padding=1):
    return nn.Sequential(
        nn.Conv2d(in_channels, out_channels, kernel_size, stride, padding),
        nn.BatchNorm2d(out_channels),
        nn.LeakyReLU(0.2)
    )

def deconv_block(in_channels, out_channels, kernel_size=3, stride=2, padding=1, output_padding=1):
    return nn.Sequential(
        nn.ConvTranspose2d(in_channels, out_channels, kernel_size, stride, padding, output_padding),
        nn.BatchNorm2d(out_channels),
        nn.ReLU()
    )

# === Models ===
class CNNEncoder(nn.Module):
    def __init__(self, num_classes=2):
        super(CNNEncoder, self).__init__()
        self.label_emb = nn.Embedding(num_classes, 256 * 256)
        self.encoder = nn.Sequential(
            conv_block(4, 16),
            conv_block(16, 32),
            conv_block(32, 64),
            conv_block(64, 128),
            conv_block(128, 256),
            conv_block(256, 512),
            nn.AdaptiveAvgPool2d(1),
            nn.Flatten(),
            nn.Linear(512, 1),
            nn.Sigmoid()
        )

    def forward(self, img, labels):
        label_map = self.label_emb(labels).view(labels.size(0), 1, 256, 256)
        x = torch.cat((img, label_map), dim=1)
        return self.encoder(x)

class CNNDecoder(nn.Module):
    def __init__(self, latent_dim=10, num_classes=2):
        super(CNNDecoder, self).__init__()
        self.label_emb = nn.Embedding(num_classes, latent_dim)
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim * 2, 32 * 8 * 8),
            nn.ReLU(),
            nn.Unflatten(1, (32, 8, 8)),
            deconv_block(32, 64),
            deconv_block(64, 128),
            deconv_block(128, 256),
            deconv_block(256, 128),
            deconv_block(128, 16),
            nn.ConvTranspose2d(16, 3, kernel_size=3, stride=1, padding=1),
            nn.Sigmoid()
        )

    def forward(self, noise, labels):
        label_embedding = self.label_emb(labels)
        input = torch.cat((noise, label_embedding), dim=1)
        return self.decoder(input)

# === Instantiate Models and Optimizers ===
generator = CNNDecoder(latent_dim=latent_dim, num_classes=2).to(device)
discriminator = CNNEncoder(num_classes=2).to(device)
criterion = nn.BCELoss()
optimizer_D = optim.Adam(discriminator.parameters(), lr=0.0002, betas=(0.9, 0.999))
optimizer_G = optim.Adam(generator.parameters(), lr=0.0001, betas=(0.9, 0.999))

# === Image Generation Function ===
def generate_images(generator, labels, latent_dim=10, nrow=4, epoch=0):
    generator.eval()
    with torch.no_grad():
        labels_tensor = torch.tensor(labels, device=device)
        z = torch.randn(len(labels), latent_dim, device=device)
        fake_images = generator(z, labels_tensor).cpu()
        grid = vutils.make_grid(fake_images, nrow=nrow, normalize=True, value_range=(0, 1))

        plt.figure(figsize=(nrow * 2, 2.5))
        plt.axis("off")
        plt.title(f"Generated Images (Epoch {epoch})")
        plt.imshow(grid.permute(1, 2, 0))
        plt.show()

# === Progressive Growing ===
def progressive_growing(generator, discriminator, train_loader, epochs, latent_dim, device):
    resolutions = [ 64, 128, 256]
    for step, res in enumerate(resolutions):
        print(f"\n--- Training at resolution {res}x{res} ---")
        train_one_stage(res, step, generator, discriminator, train_loader, epochs, latent_dim, device)

# === Training One Stage ===
def train_one_stage(resolution, step, generator, discriminator, train_loader, epochs, latent_dim, device):
    generator.train()
    discriminator.train()

    for epoch in range(epochs):
        for i, (real_images, labels) in enumerate(train_loader):
            real_images, labels = real_images.to(device), labels.to(device)
            batch_size = real_images.size(0)

            # === Train Discriminator ===
            optimizer_D.zero_grad()
            real_labels = torch.ones(batch_size, 1, device=device)
            fake_labels = torch.zeros(batch_size, 1, device=device)

            real_outputs = discriminator(real_images, labels)
            real_loss = criterion(real_outputs, real_labels)

            z = torch.randn(batch_size, latent_dim, device=device)
            fake_images = generator(z, labels)
            fake_outputs = discriminator(fake_images.detach(), labels)
            fake_loss = criterion(fake_outputs, fake_labels)

            d_loss = real_loss + fake_loss
            d_loss.backward()
            optimizer_D.step()
            d_losses.append(d_loss.item())  # Track D loss

            # === Train Generator ===
            optimizer_G.zero_grad()
            fake_images = generator(z, labels)
            fake_outputs = discriminator(fake_images, labels)
            g_loss = criterion(fake_outputs, real_labels)
            g_loss.backward()
            optimizer_G.step()
            g_losses.append(g_loss.item())  # Track G loss

        if (epoch + 1) % 40 == 0:
            labels = [0, 0, 1, 1, 0, 1, 0, 1]
            generate_images(generator, labels, latent_dim=latent_dim, nrow=4, epoch=epoch + 1)

        print(f"Epoch [{epoch+1}/{epochs}], D Loss: {d_loss.item():.4f}, G Loss: {g_loss.item():.4f}")

# === Run Training ===
progressive_growing(generator, discriminator, train_loader, 200, latent_dim, device)

# === Static Loss Plot ===
plt.figure(figsize=(12, 5))
plt.plot(d_losses, label="Discriminator Loss", color='red')
plt.plot(g_losses, label="Generator Loss", color='blue')
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.title("Training Losses")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print("âœ… Progressive Training Complete!")
```
